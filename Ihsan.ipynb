# Import Libraries And Dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from tabulate import tabulate
```

```python
iris = pd.read_csv('IRIS.csv')
```

# 1) SVM With 2 features

```python
x_two_features = iris[["sepal_length", "sepal_width"]]
y = iris["species"]

x_train_two, x_test_two, y_train, y_test = train_test_split(x_two_features, y, test_size=0.2, random_state=42)

model_two_features = SVC(kernel='linear')
model_two_features.fit(x_train_two, y_train)

pred_two_features = model_two_features.predict(x_test_two)

report = classification_report(y_test, pred_two_features, output_dict=True)

print("#####################    Results with Two Features (SVM - Linear Kernel) #####################\n")
report_table = [
    [label, round(values["precision"], 2), round(values["recall"], 2), round(values["f1-score"], 2), int(values["support"])]
    for label, values in report.items() if label in y.unique()
]
print(tabulate(report_table, headers=["Class", "Precision", "Recall", "F1-Score", "Support"], tablefmt="grid"))

cm = confusion_matrix(y_test, pred_two_features)
print("\n#################### Confusion Matrix (Table) ################################\n")
cm_table = pd.DataFrame(cm, columns=["Setosa", "Versicolor", "Virginica"],
                        index=["Setosa", "Versicolor", "Virginica"])
print(tabulate(cm_table, headers="keys", tablefmt="grid"))

print("\n######################### Confusion Matrix (Matrix) ################################\n")
print(cm)

accuracy = accuracy_score(y_test, pred_two_features)
print(f"\n#################### Accuracy ######################\n\nAccuracy: {accuracy:.2f} (or {accuracy * 100:.2f}%)")
```

# 2) SVM With 4 Features

```python
x_four_features = iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
y = iris["species"]

x_train_four, x_test_four, y_train, y_test = train_test_split(x_four_features, y, test_size=0.2, random_state=42)

model_four_features = SVC(kernel='linear')
model_four_features.fit(x_train_four, y_train)

pred_four_features = model_four_features.predict(x_test_four)

report = classification_report(y_test, pred_four_features, output_dict=True)
accuracy_from_dict=report["accuracy"]
print(accuracy_from_dict)

print("######################## Results with Four Features (SVM - Linear Kernel) #################################\n")
report_table = [
    [label, round(values["precision"], 2), round(values["recall"], 2), round(values["f1-score"], 2), int(values["support"])]
    for label, values in report.items() if label in y.unique()
]
print(tabulate(report_table, headers=["Class", "Precision", "Recall", "F1-Score", "Support"], tablefmt="grid"))

cm = confusion_matrix(y_test, pred_four_features)
print("\n########################## Confusion Matrix (Table) ####################################\n")
cm_table = pd.DataFrame(cm, columns=["Setosa", "Versicolor", "Virginica"],
                        index=["Setosa", "Versicolor", "Virginica"])
print(tabulate(cm_table, headers="keys", tablefmt="grid"))

print("\n############################ Confusion Matrix (Matrix) ###############################\n")
print(cm)

accuracy = accuracy_score(y_test, pred_four_features)
print(f"\n########################### Accuracy #############################\n\nAccuracy: {accuracy:.2f} (or {accuracy * 100:.2f}%)")
```

# 3) SVM With Different Kernels

```python
x = iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
y = iris["species"]


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

kernels = ['linear', 'poly', 'rbf']

for kernel in kernels:
    model = SVC(kernel=kernel)
    model.fit(x_train, y_train)
    pred = model.predict(x_test)

    report = classification_report(y_test, pred, output_dict=True)

    print(f"\n######################## Results with {kernel.capitalize()} Kernel ########################\n")

    report_table = [
        [label, round(values["precision"], 2), round(values["recall"], 2), round(values["f1-score"], 2), int(values["support"])]
        for label, values in report.items() if label in y.unique()
    ]
    print(tabulate(report_table, headers=["Class", "Precision", "Recall", "F1-Score", "Support"], tablefmt="grid"))

    cm = confusion_matrix(y_test, pred)
    print("\n---------->>>>>> Confusion Matrix (Table) \n")
    cm_table = pd.DataFrame(cm, columns=["Setosa", "Versicolor", "Virginica"],
                            index=["Setosa", "Versicolor", "Virginica"])
    print(tabulate(cm_table, headers="keys", tablefmt="grid"))

    print("\n---------->>>>> Confusion Matrix (Matrix) \n")
    print(cm)

    accuracy = accuracy_score(y_test, pred)
    print(f"\n---------->>>>> Accuracy \n\nAccuracy: {accuracy:.2f} (or {accuracy * 100:.2f}%)")
```

# 4) Compare The Result Of All Models

#            *** #### Comparison and Observations: #####***

* Accuracy for two features vs four features: Expect higher accuracy with four features as it utilizes more information.
* Linear kernel: Performs well when data is linearly separable.
* Polynomial kernel: Handles non-linear data but may increase computation time.
* RBF kernel: Suitable for complex non-linear data, often providing good results for the Iris dataset.

```python
def evaluate_model(model, x_train, x_test, y_train, y_test, kernel_desc):
    model.fit(x_train, y_train)
    pred = model.predict(x_test)

    acc = accuracy_score(y_test, pred)

    report = classification_report(y_test, pred, output_dict=True)
    report_table = [
        [label, round(values["precision"], 2), round(values["recall"], 2), round(values["f1-score"], 2), int(values["support"])]
        for label, values in report.items() if label in y.unique()
    ]

    print(f"\n######################### Results for {kernel_desc} Kernel #################################\n")
    print(tabulate(report_table, headers=["Class", "Precision", "Recall", "F1-Score", "Support"], tablefmt="grid"))

    cm = confusion_matrix(y_test, pred)
    print("\n------------>>>>>> Confusion Matrix \n")
    cm_table = pd.DataFrame(cm, columns=["Setosa", "Versicolor", "Virginica"],
                            index=["Setosa", "Versicolor", "Virginica"])
    print(tabulate(cm_table, headers="keys", tablefmt="grid"))

    print(f"\n------------>>>>>> Accuracy \n\nAccuracy: {acc:.2f} (or {acc * 100:.2f}%)")

    results.append({"Kernel": kernel_desc, "Accuracy": acc, "Confusion Matrix": cm})

results = []

x_two_features = iris[["sepal_length", "sepal_width"]]
x_train_two, x_test_two, y_train, y_test = train_test_split(x_two_features, y, test_size=0.2, random_state=42)
evaluate_model(SVC(kernel='linear'), x_train_two, x_test_two, y_train, y_test, "Linear (2 Features)")

x_four_features = iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
x_train_four, x_test_four, y_train, y_test = train_test_split(x_four_features, y, test_size=0.2, random_state=42)

kernels = ['linear', 'poly', 'rbf']
for kernel in kernels:
    evaluate_model(SVC(kernel=kernel), x_train_four, x_test_four, y_train, y_test, f"{kernel.capitalize()} (4 Features)")

print("\n******************************* Overall Model Comparison *********************************\n")
comparison_table = [
    [result["Kernel"], f"{result['Accuracy']:.2f}"] for result in results
]
print(tabulate(comparison_table, headers=["Kernel", "Accuracy"], tablefmt="grid"))
```

# 5) Overall Accuracy calculation

```python
overall_accuracy = sum(result['Accuracy'] for result in results) / len(results)
print(f"\nOverall Average Accuracy: {overall_accuracy:.2f}")
```
